```python
from scipy.stats import multivariate_normal
import imuncertain.distribution as dist
import imuncertain.plotting.plots2D as plotting
import imuncertain.plotting.plotsND as plottingND
```

```python
rv = multivariate_normal([0.5, -0.2], [[2.0, 0.3], [0.3, 0.5]])
rv2 = multivariate_normal([1.7, -1.5], [[2.0, 0.3], [0.3, 0.5]])
d = dist.distribution(rv)
d2 = dist.distribution(rv2)
plotting.plot_samples([d, d2], 100, xlabel="X Axis Title")
plotting.plot_pdf_contour([d, d2], range_x=(-3,5), range_y=(-3,3), xlabel="X Axis Title")
```

```python
rv = multivariate_normal([0.5, -0.2, 1.0], [[1.0, 0.3, 0.1], [0.3, 0.9, 0.2], [0.1, 0.2, 1.0]])
rv2 = multivariate_normal([2.7, -2.5, 3.0], [[1.0, 0.3, 0.1], [0.3, 0.9, 0.2], [0.1, 0.2, 1.0]])
d = dist.distribution(rv)
d2 = dist.distribution(rv2)
plottingND.plot_samples([d, d2], 500)
plottingND.plot_contour([d, d2], ranges = [(-5,5), (-5,10), (-5,15)])
plottingND.plot_contour_samples([d, d2], 500, ranges = [(-5,5), (-5,5), (-5,5)])
```
